```python
#Delvin Mackenzie
#Json worldbank exercise
#Springboard

import pandas as pd
import json 
from pandas.io.json import json_normalize
from collections import Counter
import numpy as np

#I applied 2 seperate mothods to import json data for the purpose of practice even though its redundant. 

#loads worldbank json data: worldbank_json
worldbank_json = json.load((open('Downloads/data_wrangling_json/data/world_bank_projects.json')))

#loads worldbank json data into a pandas DataFrame: df
df= pd.read_json('Downloads/data_wrangling_json/data/world_bank_projects.json')

#slices countrycode column from df: country_code
country_code = df.loc[:,['countrycode']]

#creates counter on country_code: country_counter
country_counter = Counter(country_code['countrycode'].tolist())

#prints a list of tuples of the top 10 country codes and number of occurances as a list
print(country_counter.most_common(10))

#takes the json worldbank file and normalizes into a pandas DataFrame: mjtheme_normalized
mjtheme_normalized = json_normalize(worldbank_json,'mjtheme_namecode')

#Creates a counter for the 'name' column in the worldbank DataFrame: theme_counter
theme_counter = Counter(mjtheme_normalized['name'].tolist())

#prints a list of tuples of the top 10 project themes and number of occurances as a list
theme_counter.most_common(10)

```

```python
#creates pandas DataFrame from theme_counter to visually display top 10 project themes
theme_df = pd.DataFrame.from_dict(theme_counter, orient='index').reset_index().head(10)
theme_df = theme_df.rename(columns={'index':'theme', 0:'count'})
theme_df = theme_df.sort_values('count', ascending=False)
theme_df
```

```python

#creates pandas DataFrame from theme_counter to visually display top 10 countries
top_country_df = pd.DataFrame.from_dict(country_counter, orient='index').reset_index().head(10)
top_country_df = top_country_df.rename(columns={'index':'theme', 0:'count'})
top_country_df = top_country_df.sort_values('count', ascending=False)
top_country_df
```
